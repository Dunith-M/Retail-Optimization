# Importing Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 20)
```

#Load Data

```python
df = pd.read_csv("../data/raw/sales_train_validation.csv")
```

```python
print("Rows, Columns:", df.shape)
```

```python
df.head()
```

```python
df.info()
```

#Count Unique

```python
print("Unique Products:", df["item_id"].nunique())
print("Unique Departments:", df["dept_id"].nunique())
print("Unique Categories:", df["cat_id"].nunique())
print("Unique Stores:", df["store_id"].nunique())
print("Unique States:", df["state_id"].nunique())
```

#about Sales Columns 

```python
sales_cols = [col for col in df.columns if col.startswith("d_")]

print("Total Daily Columns:", len(sales_cols))
```

```python
df["total_sales"] = df[sales_cols].sum(axis=1)

df[["id", "item_id", "store_id", "total_sales"]].head()
```

#No Sales

```python
zero_sales = df[df["total_sales"] == 0]

print("Zero Sales Rows:", zero_sales.shape[0])

zero_sales[["id", "item_id", "store_id"]].head(20)
```

# Top 10 Selling Products & Visualization

```python
top_products = (
    df.groupby("item_id")["total_sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_products)
```

```python
plt.figure(figsize=(12,6))
top_products.plot(kind="bar")
plt.title("Top 10 Selling Products")
plt.ylabel("Total Units Sold")
plt.xticks(rotation=45)
plt.show()
```

# Store Count and Visualization

```python
store_counts = df["store_id"].value_counts()

print(store_counts)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="store_id", order=df["store_id"].value_counts().index)
plt.title("Rows per Store")
plt.xticks(rotation=45)
plt.show()
```

# Category Counts and Visualization

```python
cat_counts = df["cat_id"].value_counts()

print(cat_counts)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="cat_id", order=df["cat_id"].value_counts().index)
plt.title("Rows per Category")
plt.show()
```

# No of States

```python
print(df["state_id"].value_counts())
```

# Average Sales Per Store

```python
avg_store_sales = (
    df.groupby("store_id")["total_sales"]
    .mean()
    .sort_values(ascending=False)
)

print(avg_store_sales)
```

# Products with Highest Zero Days

```python
df["zero_days"] = (df[sales_cols] == 0).sum(axis=1)

df[["item_id", "store_id", "zero_days"]].sort_values(
    by="zero_days",
    ascending=False
).head(10)
```
