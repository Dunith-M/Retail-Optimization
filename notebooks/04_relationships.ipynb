# Identify ID Columns and Sales Columns

```python
id_cols = ["id", "item_id", "dept_id", "cat_id", "store_id", "state_id"]
sales_cols = [col for col in sales.columns if col.startswith("d_")]

print("ID Columns:", id_cols)
print("Number of Sales Day Columns:", len(sales_cols))
print("First 5 Sales Columns:", sales_cols[:5])
print("Last 5 Sales Columns:", sales_cols[-5:])
```

Sales

```python
sales_sample = sales.head(5)

sales_sample
```

```python
# =========================
# CELL 8 — Convert Sales Wide Format to Long Format
# =========================

sales_long = sales_sample.melt(
    id_vars=id_cols,
    value_vars=sales_cols,
    var_name="d",
    value_name="sales"
)

sales_long.head(20)
```

```python
print("Original Sample Shape:", sales_sample.shape)
print("Long Format Shape:", sales_long.shape)
```

# Understand One Product Sales Over Time

```python
one_product = sales_long[sales_long["id"] == sales_long["id"].iloc[0]]

one_product.head(20)
```

# Prepare Calendar Columns

```python
calendar["date"] = pd.to_datetime(calendar["date"])

calendar_small = calendar[
    [
        "d",
        "date",
        "wm_yr_wk",
        "weekday",
        "wday",
        "month",
        "year",
        "event_name_1",
        "event_type_1",
        "snap_CA",
        "snap_TX",
        "snap_WI"
    ]
]

calendar_small.head()
```

```python
sales_calendar = sales_long.merge(
    calendar_small,
    on="d",
    how="left"
)

sales_calendar.head(20)
```

```python
print("Sales Long Shape:", sales_long.shape)
print("Sales + Calendar Shape:", sales_calendar.shape)

sales_calendar.isnull().sum()
```
